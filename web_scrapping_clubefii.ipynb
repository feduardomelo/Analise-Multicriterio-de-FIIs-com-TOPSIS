```python
lista_indicadores_fii = []
for fii in fundos_ifix:
# for fii in teste:
    #Instalando o driver
    servico = Service(ChromeDriverManager().install())

    #Abrindo o navegador
    navegador = webdriver.Chrome(service=servico)
    #Criando url
    url = f"https://www.clubefii.com.br/fiis/{fii}"

    
    try:
        #Abrindo o navegador
        navegador.get(url)

        time.sleep(1)
    
        #Coletando infos
        # tx_adm = navegador.find_element(By.XPATH,'//*[@id="page-fundo-basico"]/table[1]/tbody/tr[3]/td[2]/span').text
        tx_adm = navegador.find_element(By.XPATH,'//*[@id="page-fundo-basico"]/table[1]/tbody/tr[2]/td[2]/span').text
        vacancia = navegador.find_element(By.XPATH, '//*[@id="vacancia"]/div[2]/div[1]/strong[2]').text
        # //*[@id="page-fundo-basico"]/table[1]/tbody/tr[2]/td[2]/span
        # //*[@id="page-fundo-basico"]/table[1]/tbody/tr[3]/td[2]/span
        time.sleep(1)

        #Criando dict
        dicionario = {  "fii": fii, 
                        "vacancia": vacancia,
                        "tx_adm": tx_adm

                        }

        #Criando lista de dicts
        lista_indicadores_fii.append(dicionario)
    except Exception as e:
        print(e)

    #Fecha o navegador
    navegador.quit()

```

```python
df = pd.DataFrame.from_dict(lista_indicadores_fii)
```
